# O que Credit Scoring?

- Estimar a probabilidade de um evento acontecer frente a uma base de dados
- Probabilidade : 0 ou 1 
- Diversas aplicações

# O que é Credit Scoring - Concessão

- Evento : um tomador de crédito se tornar inadimplemente
- Classificação dos solicitantes : evento binário ( 0, 1)
- Toda  concessão de crédito está intrinsecamente relacionada ao risco de inadimplência


# Risco de Crédito

- Probabilidade da expectativa de receber um montante de dinheiro em um determinado período de tempo, não se concretize
- Parte dos juros associados vem do risco que é assumido ao emprestar determinada quantia
